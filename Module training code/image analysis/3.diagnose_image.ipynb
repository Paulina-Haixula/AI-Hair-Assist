# Load Model + Diagnose Uploaded Image

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import os
```

```python
# Load model
model = load_model("hair_disease_classifier.h5")

# Disease labels (match directory names used in training)
class_labels = [
    'Alopecia Areata',
    'Contact Dermatitis',
    'Folliculitis',
    'Head Lice',
    'Lichen Planus',
    'Male Pattern Baldness',
    'Psoriasis',
    'Seborrheic Dermatitis',
    'Telogen Effluvium',
    'Tinea Capitis'
]
# Load image
img_path = '0026.jpg'
img = load_img(img_path, target_size=(224, 224))
img_array = img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)
```

```python
# Predict
prediction = model.predict(img_array)
binary_result = (prediction > 0.5).astype(int)[0]

# Output DataFrame
df_result = pd.DataFrame([binary_result], columns=class_labels)
display(df_result)
```
